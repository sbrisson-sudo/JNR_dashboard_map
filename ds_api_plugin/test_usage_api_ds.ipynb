```python
type_to_parser = {
    "IntegerNumberChamp" : lambda i : 0 if i=="" else int(i),
    "CheckboxChamp" : lambda b : b == "true"
}


def nodes_to_table_corres(nodes_data, key="label", value="stringValue", parse_type=True):
    """
    Transforme une liste de nodes en une table de correspondance label : valeur
    """
    if parse_type : 
        dict_data = {}
        for node in nodes_data :
            try : 
                dict_data[node[key]] = type_to_parser[node["__typename"]](node[value])
            except KeyError: # type not referenced
                dict_data[node[key]] = node[value]
        return dict_data
    else : 
        return {node[key] : node[value] for node in nodes_data}

def get_node_by_attribute(nodes_data, key, value):
    for node in nodes_data:
        try:
            if node[key] == value:
                return node
        except KeyError:
            pass
    return None
```
