```python
import pandas as pd
```

```python
in_file = "../in_data/dossiers_journee-nationale-de-la-resilience-appel-a-projets_2023-08-01_16-59.xlsx"

df_dossier = pd.read_excel(in_file, sheet_name="Dossiers")
df_action_1 = pd.read_excel(in_file, sheet_name="(3344502) Action") # actions ayant lieu qu'une fois
df_action_2 = pd.read_excel(in_file, sheet_name="(3308253) Action") # actions ayant lieu plusieurs fois
df_loc_action2 = pd.read_excel(in_file, sheet_name="(3308215) Localisation")

# retirer les lignes vides (champ non d'action non rempli)

df_action_1 = df_action_1[~df_action_1["Nom de l'action"].isna()]
df_action_2 = df_action_2[~df_action_2["Nom de l'action"].isna()]
df_loc_action2 = df_loc_action2[~df_loc_action2["Nom de l'action"].isna()]
```

```python
# Pour chaque dossier on compte combien de mentions sont faites dans chacun des 3 onglets
df = pd.DataFrame(
    index=df_dossier["ID"],
    columns = ["Action 1", "Action 2", "Localisation"]
    )

for c,df_ in zip(["Action 1", "Action 2", "Localisation"], [df_action_1, df_action_2, df_loc_action2]):
    df[c] = df_["Dossier ID"].value_counts().rename(c)
    df = df.fillna(0)
    df[c] = df[c].astype(int)

# ajouter état du dossier
df = pd.merge(
    df.reset_index(), 
    df_dossier[
        [
            "ID","État du dossier",
            "1. Une action se déroule dans une seule localisation ou est dématérialisée",
            "2 - Une ou plusieurs de vos actions se déroulent dans plusieurs localisations"
        ]], how="left", on="ID")
```

```python
# Dossiers problématiques : plusieurs actions se déroulant dans plusieurs localisation
# Problème > impossible de lier les info sur l'action (risques traités / public ciblé) à la localisation
# Solution : 
# - Chercher des matchs entre les noms d'action (fonctionnel si nom identique, sinon compliqué)
# - Modifier le formulaire démarches administratives

for dossier_id in df[(df["Action 2"]>1) & (df["Action 2"] != df["Localisation"])]["ID"] :
    print(dossier_id)
    print("Action 2 :", df_action_2[df_action_2["Dossier ID"]==dossier_id]["Nom de l'action"].to_list())    
    print("Localisation :", df_loc_action2[df_loc_action2["Dossier ID"]==dossier_id]["Nom de l'action"].to_list())
```
